# Webscraping ______________________

```python
### imports
import requests
import io
import pandas as pd
import re

from bs4 import BeautifulSoup as bs
```

```python
### config
URL = "https://ryman.com/events/"
```

```python
### request url and let bs work
response = requests.get(URL)
soup = bs(response.text, "html.parser")

#'eventItem entry'
# cant print huge stuff in notebook w/o changing settings
#print(soup.prettify())
```

```python
### take in each card that holds our needed info
### each card contains the class *eventItem*
cards = soup.find_all("div", class_="eventItem")

rows = []

for card in cards:
    def get_text_or_none(tag):
        return tag.get_text(" ", strip=True) if tag else None

    day = get_text_or_none(card.find("span", class_="m-date__day"))
    month = get_text_or_none(card.find("span", class_="m-date__month"))
    year = get_text_or_none(card.find("span", class_="m-date__year"))
    hour = get_text_or_none(card.find("span", class_="m-date__hour"))
    location = get_text_or_none(card.find("div", class_="location"))
    presented_by = get_text_or_none(card.find("div", class_="presented-by"))
    featuring = get_text_or_none(card.find("h4", class_="tagline"))
    
    title_tag = card.find("h3")
    title = get_text_or_none(title_tag)

    date = " ".join(x for x in [month, day, year] if x)

    rows.append({
        "title": title,
        "featuring": featuring,
        "date": date,
        "time": hour,
        "location": location,
        "presented_by": presented_by,
    })

for row in rows[:5]:
    print(row)
    
```

```python
### print our df
ryman_df = pd.DataFrame(rows)
ryman_df.tail(50)
```
